```python
import os
import re
from PIL import Image
import pickle
import numpy as np
import cv2
import gc
from tqdm.notebook import tqdm

shape = 800

path_save = "/content/gdrive/MyDrive/Progetti Personali/foto_ale/save/file"
path_original = "/content/gdrive/MyDrive/Progetti Personali/foto_ale/Foto All"
path_list = '/content/gdrive/MyDrive/Progetti Personali/foto_ale/save/list_file/'

image_pattern = r".*\.(jpg|jpeg|png)"
```

```python
original_photo_path = [os.path.join(path, f) for path, subdirs, files in os.walk(path_original)  for f in files if re.match(image_pattern, f, flags=re.I)]
print(len(original_photo_path))
```

```python
try:
  with open(os.path.join(path_list, 'list_image.pkl'), 'rb') as file:
    list_of_image = pickle.load(file)
  
except:
  list_of_image = []

print(len(list_of_image))
```

```python
filtered_photo_path = [x for x in original_photo_path if x in list_of_image]
```

```python
for i, file_path in tqdm(enumerate(filtered_photo_path)):
  image_ = Image.open(file_path)
  image_.save(os.path.join(path_save, f'saved_file_{i}.png'))
```
